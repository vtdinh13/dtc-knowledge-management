Checks whether the filename matches this exact pattern:

- ^ means start of the string
- llm_response_ matches that literal prefix
- (.+) captures one or more characters
- \.json matches the literal text .json
- $ means end of the string

```python
import re

match = re.match(r"^llm_response_(.+)\.json$", test_doc_path.name)
id = match.group(1) if match else None

print(id)  
```

```python
import json
```

```python
with open(test_doc_path, "r", encoding="utf-8") as f:
    test_doc = json.load(f)
```

```python
from yt_dlp import YoutubeDL
from pathlib import Path
import time
```

```python
youtube_link = f"https://www.youtube.com/watch?v={id}"

outpath = Path("test_videos")
outpath.mkdir(parents=True, exist_ok=True)

outpath_string = str(outpath)
final_dest = f"{outpath_string}/{id}.mp4"

start = time.perf_counter()
with YoutubeDL({"outtmpl": final_dest, "format": "mp4"}) as ydl:
    ydl.download([youtube_link])
end = time.perf_counter()

duration=end-start
```

```python
print(f"Video took {duration} seconds to download.")
```
