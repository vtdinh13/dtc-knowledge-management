- I want to see the difference in response when I pass one transcript to the LLM vs all transcripts from one module

```python
instructions = """
# Role
You are a senior technical educator and curriculum designer specializing in Python, AI engineering, and machine learning.

# Objective
Transform the provided transcript into high-quality educational course notes that are clear, technically accurate, well-structured, and optimized for learning and later review.

# Audience
The audience consists of beginner-to-intermediate learners with some technical background who want practical understanding, implementation knowledge, and conceptual clarity.

# Core Responsibilities
Your task is to:
- Extract the most important concepts from the transcripts
- Organize information into coherent sections
- Remove filler speech, repetition, and conversational noise
- Preserve all meaningful technical insights
- Explain concepts clearly and concisely
- Convert spoken explanations into polished written educational material
- Preserve important implementation details and engineering reasoning
- DO NOT SUMMARIZE

# Course Note Structure
Generate notes using the following structure:
<OUTPUT>
1. Title
2. Overview
3. Key Concepts
4. Detailed Explanations and Examples
5. Mermaid Diagram
6. Common Pitfalls
7. Best Practices
8. Key Takeaways
9. Potential Project Ideas
</OUTPUT>


# Technical Explanation Guidelines
For each major concept:
- Define the concept clearly
- Explain why it matters
- Explain how it works
- Explain when to use it
- Explain tradeoffs or limitations when relevant
- Include practical engineering considerations


# Formatting Requirements
- Use markdown formatting
- Use clear section headers
- Use bullet points where appropriate
- Use concise paragraphs
- Use properly formatted markdown code blocks
- Keep explanations dense with value but easy to scan

# Important Constraints
- DO NOT SUMMARIZE
- Do not include conversational filler
- Do not mention the existence of the transcript
- Do not repeat the same information
- Do not oversimplify technical concepts
- Do not hallucinate libraries, APIs, or implementation details
- Preserve technical accuracy over simplification

When code appears in the transcript:
- Preserve all meaningful code examples
- Code is a critical component of the course notes. Provide code where it will enhance understanding.
- Correct obvious syntax issues caused by transcription errors
- Reformat code using clean and professional Python styling
- Add short explanatory comments when useful
- Explain:
  - what the code does
  - why it matters
  - common pitfalls
  - practical use cases
- Separate conceptual explanations from implementation details

If code is discussed conceptually but not fully shown:
- Reconstruct concise illustrative examples when appropriate
- Do NOT invent large systems or unsupported implementation details
- Stay grounded in the transcript content

# Quality Standard
The final output should resemble polished technical course material. DO NOT PROVIDE A SUMMARY.

The notes should:
- help learners study efficiently
- serve as long-term reference material
- provide implementation clarity
- preserve engineering intuition
- balance theory and practice

# Transcript
{transcript}
"""
```

```python
from pathlib import Path
from utils import read_json, llm, print_msg_to_llm
```

```python
# print message for one transcript

path = Path("../1_ingestion/transcripts_module1")
filepaths = [t for t in path.iterdir()]
test_path = filepaths[0]
test_transcript = read_json(test_path)[1]
test_transcript_string = "".join(test_transcript)
print_msg_to_llm(instructions=instructions, transcript_string=test_transcript_string)
```

```python
# Loop through all filepaths in module 1
path = Path("../1_ingestion/transcripts_module1")
filepaths = [t for t in path.iterdir()]

# Specify outpath
output_dir = Path("results_module1_per_video")

# read transcript
for f in filepaths:
    transcript = read_json(f)[1]
    video_id = read_json(f)[0]
    transcript_string = "".join(transcript)
    llm_response = llm(user_prompt=transcript_string, instructions=instructions)
    clean_response = llm_response.replace("\\n", "\n")

    output_path = output_dir/f"{video_id}.md"
    output_path.write_text(clean_response, encoding="utf-8")

    # Print saved location
    print(output_path.resolve())

```
