```python
import plotly.express as px
import pandas as pd
import json
```

```python
with open ("ml_zoomcamp_metadata.json", "r") as f:
    metadata = json.load(f)

dat = []
for i in metadata:
    dat.append(
        {"id": i["id"],
        "title": i["title"],
        "duration": i["duration"],
        "view_count": i["view_count"]}

    )

df = pd.DataFrame(dat)

# clean title
remove_phrases = ["ML Zoomcamp 2025", "ML Zoomcamp "]
for i in remove_phrases:
    df["cleaned_title"] = df["title"].str.replace(i, "", regex=False)

# separate by modules
df["module"] = df["cleaned_title"].str.extract(r"^(\d+)")

# fix the first two modules
df.loc[[0, 1], "module"] = ["0", "0"]

# replace NaN - the kubernetes video with 100
df["module"] = df["module"].fillna(11)

# change module column into int
df["module"] = df["module"].astype(int)
df.head(5)
```

```python
duration = px.box(
    data_frame=df, 
    x="module",
    y="duration",
    color="module",
    hover_name="title"
)
duration.show()
```

```python
view_count = px.box(
    data_frame=df,
    x="module",
    y="view_count",
    color="module",
    hover_name="title",
)
view_count.show()
```

```python
corr = px.scatter(
    data_frame=df,
    x="duration",
    y="view_count",
    hover_name="title",
    hover_data={
        "duration": True,
        "view_count": True,
        "title": False
    }
)
corr.show()
```
