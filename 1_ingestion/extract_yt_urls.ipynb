```python
import yt_dlp
```

```python
playlist_url = "https://www.youtube.com/playlist?list=PL3MmuxUbc_hIhxl5Ji8t4O6lPAOpHaCLR"

ydl_opts = {
        "quiet": True,
        "extract_flat": True,
        "skip_download": True,
    }

yt = yt_dlp.YoutubeDL(ydl_opts)

yt_dict = yt.extract_info(playlist_url, download=False)


yt_dict
```

```python
type(yt_dict["entries"])
```

```python
yt_dict["entries"][0]
```

```python
video_metadata = []

for i in yt_dict["entries"]:
    
    temp_dict = {}

    for key,value in i.items():
        if key in ["title", "url", "id", "views", "duration", "view_count"]:
            temp_dict[key] = value
    video_metadata.append(temp_dict)

video_metadata
```

```python
import json

with open("ml_zoomcamp_metadata.json", "w") as f:
    json.dump(video_metadata, f, indent=4)
```
